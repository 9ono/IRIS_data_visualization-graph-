# 0. data_load

```python
import pandas as pd
df = pd.read_csv('./iris.csv')
df.head()
```

## 1. mean_median_standard-deviation

```python
mms_df = pd.DataFrame(columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], index = ['mean', 'median', 'std'])
```

```python
for col in mms_df.columns.values:
    mms_df[col].loc['mean'] = df[col].mean()
    mms_df[col].loc['median'] = df[col].median()
    mms_df[col].loc['std'] = df[col].std()
mms_df.head(len(mms_df))
```

## 2. correlation

```python
corr = df.corr()
corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## 3-1. percentile

```python
for col in mms_df.columns.values:
    plt.plot(np.percentile(df[col], range(0, 100)))
    plt.title(col)
    plt.grid()
    plt.xlabel('percentile')
    plt.show()
```

## 3-2. boxplot

```python
s_list = []
for col in mms_df.columns.values:
    s_list.append(df[col])
plt.boxplot(s_list, sym="b*")
plt.grid()
plt.xticks([1, 2, 3, 4], mms_df.columns.values)
plt.show()
```

## 3-3. histogram

```python
for col in mms_df.columns.values:
    plt.title(col)
    plt.hist(df[col], bins = 10)
    plt.grid()
    plt.show()
```

```python
for col in mms_df.columns.values:
    try:
        sns.distplot(df[col], rug=True, bins = 10)
        plt.grid()
        plt.show()
    except TypeError:
        pass
```

## 4. catplot

```python
for col in mms_df.columns.values:
    for col_2 in mms_df.columns.values:
        if col == col_2:
            break
        sns.catplot(x = col, y = col_2, hue = 'species', kind = 'bar', data = df, height = 5, aspect = 8)
        plt.title('                                         [' + col + '] and [' +  col_2 + ']')
        plt.show()
        print()    #
```

## 5. pairplot

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"], size=3)
plt.show()
```

## 6. jointplot

```python
for col in mms_df.columns.values:
    for col_2 in mms_df.columns.values:
        if col == col_2:
            break
        sns.jointplot(x = col, y = col_2, data = df)
        plt.title('                                         [' + col + '] and [' +  col_2 + ']')
        plt.show()
        print()    #
```

```python
for col in mms_df.columns.values:
    for col_2 in mms_df.columns.values:
        if col == col_2:
            break
        sns.jointplot(x = col, y = col_2, data = df, kind="kde", space = 0, zorder = 0, n_levels = 6)
        plt.title('                                         [' + col + '] and [' +  col_2 + ']')
        plt.show()
        print()    #
```
